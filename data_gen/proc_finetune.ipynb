```python
import pandas as pd
import jsonlines
import numpy as np
import sklearn.metrics
```

```python
df = pd.read_csv("results/npi/npi_finetune_0_test.tsv", sep='\t')
df.head()
```

```python
def untokenize(sentence):
    sentence = sentence.replace(" ##", "")
    sentence = sentence.replace(" ' ", "'")
    sentence = sentence.replace(" &apos;", "'")
    return sentence
df["sentence_1"] = df["sentence_1"].apply(untokenize)
```

```python
with jsonlines.open(f"npi/npi_test.jsonl") as reader:
    data = [obj for obj in reader]
```

```python
def get_co_occur(sentence):
    for sent in data:
        # TODO: some of the sentences look incomplete for some reason (22/750)
        if sentence in sent["sentence"]:
            return sent["co-occurs"]
    print(sentence)
    print("Error!")
    
def get_dist(row):
    if row["co_occurs"]:
        return row["true_label"]
    else:
        return 0 if row["true_label"] == 1 else 0
    
df["co_occurs"] = df["sentence_1"].apply(get_co_occur)
df["dist"] = df.apply(get_dist, axis=1)
df.head()
```

```python
def get_error(row):
    if row["prediction"] == row["true_label"]:
        return 0
    else:
        return 1

df["error"] = df.apply(get_error, axis=1)
df.head()
```

```python
both = df[df["true_label"] == 1][df["co_occurs"] == 1]
neither = df[df["true_label"] == 0][df["co_occurs"] == 1]
true_only = df[df["true_label"] == 1][df["co_occurs"] == 0]
distractor_only = df[df["true_label"] == 0][df["co_occurs"] == 0]

filters = [(df, "Test"), (both, "Both"), (neither, "Neither"), (true_only, "True only"), (distractor_only, "Distractor only")]
for (data, name) in filters:
    print(name)
    print("Error: {}".format(str(np.sum(data["error"]) / len(data))))
    print()
```

```python
print("I(pred;true): {}".format(sklearn.metrics.mutual_info_score(df["true_label"], df["prediction"])))
print("I(pred;dist): {}".format(sklearn.metrics.mutual_info_score(df["dist"], df["prediction"])))
```
